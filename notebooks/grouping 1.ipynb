**Listing 1: Grafische Darstellung der Klangereignisse/-folgen mit Audio-Player:**

```python
def plotEventWithPlayer(events,showProximity=0,showDurationRatio=0,showSequenceEnd=0,t=1.5):

    samples = audio[events[0]["start"]:events[len(events)-1]["end"]]
    startsample = events[0]["start"]
    
    t_audio = np.linspace(0, len(samples)/samplerate, num=len(samples))
    textpositionY = samples.max()
    
    colors = cm.rainbow(np.linspace(0, 1, 5))
    
    fig = plt.figure(figsize=(18, 12), dpi=400)
    plt.xlabel('Zeit (Sekunden)')
    plt.plot(t_audio,samples,'0.5',label="Audio Waveform",linewidth=0.9)

    for i in range (0,len(events)):
        markBegin = (events[i]["start"]-startsample)/samplerate
        markEnd = (events[i]["start"]-startsample+events[i]["features"]["effLength"])/samplerate
        eventEnd = (events[i]["start"]-startsample + (events[i]["end"]-events[i]["start"]))/samplerate
        plt.axvline(x=markBegin,linewidth=2,color='r')

        if showDurationRatio:
            plt.text(markEnd,textpositionY,'%1.2f'%events[i]["features"]["durationRatio"],bbox=dict(boxstyle='round', facecolor=colors[0], alpha=0.5),fontsize=14,horizontalalignment='right',)
            plt.axvspan(markBegin,markEnd, color='yellow', alpha=0.3)
            
        if showProximity & (i>0):
            plt.text(markEnd,textpositionY,'%1.2f'% proximity(i,events),bbox=dict(boxstyle='round', facecolor=colors[1], alpha=0.5),fontsize=14,horizontalalignment='right',)

        if showSequenceEnd:
            if sequenceEnd(i,events,t=t):
                plt.axvline(x=eventEnd,linewidth=10,color=colors[2],alpha=0.5)

    if showDurationRatio:
        plt.plot(0,0,color=colors[0], label='duration ratio',linewidth=10)

    if showProximity:
        plt.plot(0,0,color=colors[1], label='proximity',linewidth=10)
        
    if showSequenceEnd:
        plt.plot(0,0,color=colors[2], label='sequence end mark',linewidth=10)        
        
    plt.legend(loc=4)
            
    wavPlayer(samples,samplerate)
```

### Variante 1: Duration Ratio

Jedes Klangereignis besteht aus einem hörbarem Bereich (gelb markierter Bereich) und einer Pause unterhalb der Wahrnehmungsschwelle von -90db, die den Abstand zum nächsten hörbaren Klang markiert. Zum direkten Vergleich dieser Werte zwischen mehreren Klangereignissen wir das Verhältnis (*durationRatio, dR*) des ersten Teils (*effectiveDuration*) zur Gesamtlänge (*duration*) berechnet:

$$durationRatio(i)=\frac{effectiveDuration_i}{duration_i}$$

**Beispiel 1:**

```python
plotEventWithPlayer(se[15:16],showDurationRatio=1)
```

**Beispiel 2:**

```python
plotEventWithPlayer(se[45:53],showDurationRatio=1)
```

Je kleiner die *duration ratio*, desto länger ist die Pause bis zum nächsten Klangereignis. Durch unterschiedlich lange Ausklingphasen ist der Wert zum Bestimmen zeitlicher Struktur zu ungenau, kann jedoch als Maß der rhythmischen Unterteilung für weitere Analyseansätze dienen.

### Variante 2: Proximity

Das Verhältnis zwischen dem zeitlichen Abstands eines Klangereignisses von nächsten und dem zum vorherigen Klangereignisses  kann die zeitliche Nähe unabhängig von der Dauer des tatsächlichen Klangs bestimmt werden:

$$proximity(i)=\frac{duration_i}{duration_{i-1}}$$


Durch Einfügen eines *ritardando-Faktors* bleibt der Quotient auch bei stetigem Verlangsamen des Tempos noch kleiner 1:


$$proximity(i)=\frac{duration_i}{duration_{i-1} * ritardFactor}$$

Wenn

$$proximity(i) < 1$$

liegt das betrachtete Klangereignis zeitlich näher am folgenden als am vorausgehenden Klangereignis.



```python
def proximity(i,events,ritardFactor = 1.15):
    if (i==0) | (i >= len(events)):
        return 0
    proximity =  float(events[i]["end"] - events[i]["start"])   / float(events[i-1]["end"] - events[i-1]["start"]) * ritardFactor
    return proximity
```

**Beispiel 3a:**

```python
plotEventWithPlayer(se[45:53],showProximity=1)
```

Solange *proximity* einen bestimmten Schwellenwert *t* nicht übersteigt oder zunimmt wird eine zusammenhängende Klangfolge angenommen:

$$sequenceEnd(i)=\left\{\begin{matrix} 
 0 & \text{falls }  proximity(i) < t\\
 0 & \text{falls }  proximity(i) < proximity(i+1)\\
 1 & \text{falls }  proximity(i) \geq  t\\
\end{matrix}\right.$$

```python
def sequenceEnd(i,events,t=1.5):
    if proximity(i,events)<t:
        return 0
    if proximity(i,events)<proximity(i+1,events):
        return 0
    if proximity(i,events)>= t:
        return 1
    return 1
```

```python
plotEventWithPlayer(se[45:53],showProximity=1,showSequenceEnd=1)
```

Das Beispiel 2b zeigt, dass bei Abnahme des *proximity*-Werts um einen im Verhältnis kleinen Anteil *s* (hier 34.77% von Klangereignis 5 auf 6) weiterhin eine zusammenhängende Folge vorliegen kann:

$$sequenceEnd(i)=\left\{\begin{matrix} 
 0 & \text{falls }  proximity(i) < t\\
 0 & \text{falls }  proximity(i) * (1-s) < proximity(i+1)\\
 1 & \text{falls }  proximity(i) \geq  t\\
\end{matrix}\right.$$

```python
def sequenceEnd(i,events,t=1.5,s=0.35):
    if proximity(i,events)<t:
        return 0
    if proximity(i,events) * s < proximity(i+1,events):
        return 0
    if proximity(i,events)>= t:
        return 1
    return 0
```

**Beispiel 4c** mit **t=2.0** und **s=0.35**:

```python
plotEventWithPlayer(se[65:86],showProximity=1,showSequenceEnd=1,t=2.0)
```

**Beispiel 4d** mit **t=2.25** und **s=0.35**:

```python
plotEventWithPlayer(se[65:86],showProximity=1,showSequenceEnd=1,t=2.25)
```
