## <br> Variante 1: Duration Ratio

Jedes Klangereignis besteht aus einem hörbarem Bereich (gelb markierter Bereich) und einer Pause unterhalb der Wahrnehmungsschwelle von -90db, die den Abstand zum nächsten hörbaren Klang markiert. Zum direkten Vergleich dieser Werte zwischen mehreren Klangereignissen wir das Verhältnis (*durationRatio, dR*) des ersten Teils (*effectiveDuration*) zur Gesamtlänge (*duration*) berechnet:

$$durationRatio(i)=\frac{effectiveDuration_i}{duration_i}$$

```python
def durationRatio(i,events):
    duration = float(events[i]["end"]-events[i]["start"])
    return events[i]["features"]["effLength"]/duration
```

**Beispiel 1:**

```python
plotEventWithPlayer(se[15:16],showDurationRatio=1)
```

**Beispiel 2:**

```python
plotEventWithPlayer(se[45:53],showDurationRatio=1)
```

Je kleiner die *duration ratio*, desto länger ist die Pause bis zum nächsten Klangereignis. Durch unterschiedlich lange Ausklingphasen ist der Wert zum Bestimmen zeitlicher Struktur zu ungenau, kann jedoch als Maß der rhythmischen Unterteilung für weitere Analyseansätze dienen.

## <br> Variante 2: Proximity

Das Verhältnis zwischen dem zeitlichen Abstands eines Klangereignisses von nächsten und dem zum vorherigen Klangereignisses  kann die zeitliche Nähe unabhängig von der Dauer des tatsächlichen Klangs bestimmt werden:

$$proximity(i)=\frac{duration_i}{duration_{i-1}}$$


Durch Einfügen eines *ritardando-Faktors* bleibt der Quotient auch bei stetigem Verlangsamen des Tempos noch kleiner 1:


$$proximity(i)=\frac{duration_i}{duration_{i-1} * ritardFactor}$$

Wenn

$$proximity(i) < 1$$

liegt das betrachtete Klangereignis zeitlich näher am folgenden als am vorausgehenden Klangereignis.



```python
def proximity(i,events,ritardFactor = 1.25):
    if (i==0) | (i >= len(events)):
        return 0
    proximity =  float(events[i]["end"] - events[i]["start"]) / (events[i-1]["end"] - events[i-1]["start"]) * ritardFactor
    return proximity
```

### <br> Schwellenwertbasierte Gruppierung

**Beispiel 3a:**

```python
example3 = se[45:53]
plotEventWithPlayer(example3,showProximity=1)
```

Solange *proximity* einen bestimmten Schwellenwert *t* nicht übersteigt oder zunimmt wird eine zusammenhängende Klangfolge angenommen:

$$sequenceEnd(i)=\left\{\begin{matrix} 
 0 & \text{falls }  proximity(i) < t\\
 0 & \text{falls }  proximity(i) < proximity(i+1)\\
 1 & \text{falls }  proximity(i) \geq  t\\
\end{matrix}\right.$$

```python
def sequenceEnd(i,events,t=1.5):
    if proximity(i,events)<t:
        return 0
    if proximity(i,events)<proximity(i+1,events):
        return 0
    if proximity(i,events)>= t:
        return 1
    return 1
```

Das Beispiel 2b zeigt, dass bei Abnahme des *proximity*-Werts um einen im Verhältnis kleinen Anteil *s* (hier 34.77% von Klangereignis 9 auf 10) weiterhin eine zusammenhängende Folge vorliegen kann:

$$sequenceEnd(i)=\left\{\begin{matrix} 
 0 & \text{falls }  proximity(i) < t\\
 0 & \text{falls }  proximity(i) * (1-s) < proximity(i+1)\\
 1 & \text{falls }  proximity(i) \geq  t\\
\end{matrix}\right.$$

```python
def sequenceEnd(i,events,t=1.5,s=0.35):
    if proximity(i,events)<t:
        return 0
    if proximity(i,events) * s < proximity(i+1,events):
        return 0
    if proximity(i,events)>= t:
        return 1
    return 0
```

### <br> Variante 3: Kombination aus *duration ration* und *proximity* als zweidimensionaler Vektor (*proximityVector*)

Durch die gemeinsame Betrachung der beiden Merkmale *duration ratio* und *proximity* können die rhythmische Mikrostruktur (Unterteilung innerhalb des Klangereignisses) mit der Makrostruktur (Verhältnis der Dauer zum vorhergenden *sonic event*) in Verbindung gebracht werden.

```python
plotEventWithPlayer(se[65:186],showDurationRatio=1,showProximity=1)
```

```python
def proximityVector(i,events):
    return [durationRatio(i,events),proximity(i,events)]
```

```python
proximityVectorX=[]
proximityVectorY=[]
for i in range (65,186):
    proximityVectorX.append(proximityVector(i,se)[0])
    proximityVectorY.append(proximityVector(i,se)[1])
```

**Verteilung der Wertepaare in den Klangereignissen aus Beispiel 4:**

```python
fig = plt.figure(figsize=(10, 7), dpi=400)
plt.plot(proximityVectorX,proximityVectorY,'bo')
plt.xlabel('duration ratio')
plt.ylabel('proximity')
```

In der Verteilung erscheinen deutliche Cluster. Können diese Strukturen für die Segmentierung nutzbar gemacht werden? Als mögliches Maß kann der euklidische Abstand (*distance*) in der zweidimensionalen Fläche berechnet werden:

$$d(x,y)=\sqrt{(x_1-y_1)^2 + (x_2-y_2)^2} = \sqrt{\sum_{i=1}^{2}(x_i-y_i)^2}$$

```python
def proximityVectorDistance(i,events):
    proximityVector1 = proximityVector(i,events)
    proximityVector2 = proximityVector(i-1,events)
    return sqrt((proximityVector1[0]-proximityVector1[1])**2+(proximityVector2[0]-proximityVector2[1])**2)
```

```python
plotEventWithPlayer(se[65:156],showProximityVector=1,showProximity=1)
```
