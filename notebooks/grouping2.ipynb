## <br> Similarity-Vector

Für die Vergleichbarkeit von Klängen jenseits der zeitlichen Dimension ist es notwendig, einen gleichförmigen Timbre-Parameterraum zu schaffen. Die Forschungsgruppe *Music Perception and Cognition* am IRCAM Paris hat dazu hilfreiche Grundlagen erarbeitet (Plomp, 1970, Wessel 1979, Miller & Carterette 1975, McAdams et al., 1995), jedoch mit starkem Fokus auf die Unterscheidung verschiedener Instrumente und der Intetration subjektiver Höreindrücke (McAdams, 1999). Dies spiegelt sich auch in der verwendeten Formel für den wahrgenommenen unterschied zweier Klänge wieder:

$$d=\sqrt{k_{LAT}*LAT^2 + k_{SC} * SC^2 + k_{SS} * SS^2 + k_{SF} * SF^2}$$
<br>

<center>(ebd.)</center>

Jedes Merkmal (LAT: Log attac time, SC: spectral centroid, SS: spectral smoothness, SF: spectral flux) wird nur durch eine Dimension repräsentiert und jeweils mit einem über die ganze Datenbasis normalisierten Koeffizienten *k* multipliziert.

Da in dieser Untersuchung immer nur Klänge des gleichen Instruments verglichen werden, kann auf den Koeffizienten verzichtet werden. Um deutliche Unähnlichkeiten feststellen zu können, bedarf es aber einer höhren Auflösung je Merkmal, die aufgrund der unterschiedlichen Anzahl an Daten je nach Länge des Klangs auf eine gleiche an von Dimensionen interpoliert werden müssen. 

Eindimensionale Merkmale, die den Verlauf des gesamten Klangs beschreiben, werden entsprechend der Auflösung vervielfacht.

```python
from scipy.ndimage.interpolation import zoom

def similarityVector(event,resolution=5.0,tellfeatures=0):
    sV = np.array([])
    for feature in event["features"]:
        if (type(event["features"][feature])==type(dict())):# & (feature<>"Pitch"):
            zoomFactor = 1/(len(event["features"][feature]["raw"])/resolution)            
            sV = np.append(sV,zoom(event["features"][feature]["raw"],zoomFactor))
            if tellfeatures: print feature
    #for feature in event["features"]:
    #    if type(event["features"][feature])<>type(dict()):
    #        for i in range (0,int(resolution)):
    #            sV = np.append(sV,event["features"][feature])
    #        if tellfeatures: print feature
    return sV
```

```python
sV = similarityVector(sonicevents[0],tellfeatures=1)
```

$$similarityVector(se_i)=\begin{bmatrix}
SpectralCentroid_1 & ... & SpectralCentroid_{5} \\
Roughness _1 & ... & Roughness _{5} \\
SpectralFlux _1 & ... & SpectralFlux _{5}  \\
SpectralComplexity_1 & ... & SpectralComplexity_{5} \\
ZCR_1 & ... & ZCR_{5} \\
Pitch_1 & ... & Pitch_{5} \\
Loudness_1 & ... & Loudness_{5}  \\
SpectralRolloff_1 & ... & SpectralRolloff _{5}  \\
\end{bmatrix}$$

<br>
Um die Ähnlichkeit zwischen zwei Merkmalsvektoren zu messen, bestimmen wir zunächst deren *Unähnlichkeit* als euklidischer Abstand (Quian et al. 2004).

Im zweidimensionalen Raum: $$d(x,y)=\sqrt{(x_1-y_1)^2 + (x_2-y_2)^2} = \sqrt{\sum_{i=1}^{2}(x_i-y_i)^2}$$

für D Dimensionen:

$$d(x,y)=\sqrt{\sum_{d=1}^{D}(x_d-y_d)^2}$$

```python
def similarityMatrix(eventindex, norm=1):
    similarityVectors = []
    for event in sonicevents:
        similarityVectors.append(similarityVector(event))
    
    similarityVectorsArray = np.asanyarray(similarityVectors)
    distances = np.sqrt((((similarityVectorsArray - similarityVectorsArray[eventindex,:]))**2).sum(axis=1))
    if norm:
        distances = distances/np.amax(distances)
    sorted_distances_indices = np.argsort(distances)
    return (sorted_distances_indices,distances)
```

```python
def plotSimilarEventsWithPlayer(eventindex,audio,indexrange=[0]):

    (sorted_distances_indices,distances) = similarityMatrix(eventindex)
    
    samples = []
    eventsamples = []
    fig = plt.figure(figsize=(30, 30), dpi=400)
    subplots=[]
    
    subplotindex = 1
    
    for i in indexrange:
        eventsamples = audio[sonicevents[sorted_distances_indices[i]]["start"]:sonicevents[sorted_distances_indices[i]]["end"]]
        samples.extend(np.zeros(samplerate)) #inset one second of silence
        samples.extend(eventsamples)
    
        startsample = sonicevents[sorted_distances_indices[i]]["start"]
    
        t_audio = np.linspace(0, len(eventsamples)/samplerate, num=len(eventsamples))
        ax = plt.subplot(4,4,subplotindex)
        ax.plot(t_audio,eventsamples,'0.5',label="Audio Waveform",linewidth=0.9)
        ax.text(0.5, 0.05,'Sonic Event %i'% (sorted_distances_indices[i]+1),bbox=dict(boxstyle='round', \
                facecolor='r', alpha=0.5),fontsize=14,horizontalalignment='center',transform=ax.transAxes)
        
        ax.text(0.5, 0.9,'Similarity %2.2f'% ((1-distances[sorted_distances_indices[i]])*100.0),bbox=dict(boxstyle='round', \
                facecolor='b', alpha=0.5),fontsize=14,horizontalalignment='center',transform=ax.transAxes)       
        
        
        plt.xlabel('time (seconds)')
        subplotindex += 1
    
    wavPlayer(samples,samplerate) 
```

Es zeigt sich, dass die mathematische Distanz zwischen zwei Vektoren nicht immer mit der wahrgenommenen Ähnlichkeit übereinstimmt. Zur weiteren Segmentierung der Klänge soll dieses Ergebnis jedoch zunächst genügen.

<br>Berechnet man Ähnlichkeitsvektoren für alle Klangereignisse einer Aufnahme, lassen sich noch weitere interessante Erkenntnisse ziehen:

```python
similarityMatrices = []
maxDistance=0
for i in range(0,len(sonicevents)):
    matrix = similarityMatrix(i,norm=0)
    similarityMatrices.append(matrix)
    if np.amax(matrix) > maxDistance:
        maxDistance = np.amax(matrix)

normedSimilarityMatrices = []
for matrix in similarityMatrices:
    normedSimilarityMatrices.append(matrix/maxDistance)
```

```python
firstDistance = [normedSimilarityMatrices[i][1][normedSimilarityMatrices[i][0][1]] for i in range(0,len(normedSimilarityMatrices))]
sortedFirstDistances = np.argsort(firstDistance)
```

### <br> Schwellenwertbasierte Gruppierung

```python
def plotSimilarityWithPlayer(events,audio,showSimilarity=0,showSequenceEnd=0,t=0.5):

    samples = audio[events[0]["start"]:events[len(events)-1]["end"]]
    startsample = events[0]["start"]
    
    t_audio = np.linspace(0, len(samples)/samplerate, num=len(samples))
    
    #create different colors for feature distinction
    colors = cm.rainbow(np.linspace(0, 1, 7))
    
    fig = plt.figure(figsize=(35, 20), dpi=400)
    ax1 = fig.add_subplot(221)
    plt.xlabel('time (seconds)')
    ax1.plot(t_audio,samples,'0.5',label="Audio Waveform",linewidth=0.9)
    trans = transforms.blended_transform_factory(ax1.transData, ax1.transAxes)
            
    #prepare similarity
    if showSimilarity | showSequenceEnd:
        Similarity = []
        SimilarityTimestamp = []
        for i in range (0,len(events)):
            SimilarityTimestamp.append((events[i]["start"]-startsample)/samplerate)
            Similarity.append(similarity(i,sonicevents))    
            
    ax2=plt.twinx()
    
    for i in range (0,len(events)):
        markBegin = (events[i]["start"]-startsample)/samplerate
        markEnd = (events[i]["start"]-startsample+events[i]["features"]["effLength"])/samplerate
        eventEnd = (events[i]["start"]-startsample + (events[i]["end"]-events[i]["start"]))/samplerate
        ax1.axvline(x=markBegin,linewidth=2,color='r')
        ax1.text(markBegin+0.1, 0.05,'%i'% (i+1),bbox=dict(boxstyle='round', \
                facecolor='r', alpha=0.5),fontsize=14,horizontalalignment='left',transform=trans)
            
        if showSimilarity & (i>0):
            plt.text(markEnd,0.9,'%1.2f'% similarity(i,events),bbox=dict(boxstyle='round', \
                    facecolor=colors[5], alpha=0.5),fontsize=14,horizontalalignment='right',transform=trans)
            ax2.plot(SimilarityTimestamp,Similarity,color=colors[5],marker='o',lw=0,ms=10)
            ax2.set_ylabel("similarity",color=colors[5])
            if showSequenceEnd:
                if sequenceEnd(i,events,t=t):
                    plt.axvline(x=eventEnd,linewidth=10,color=colors[3],alpha=0.5)
                    
    if showSimilarity:
        plt.plot(0,0,color=colors[5], label='similarity',linewidth=10)

    if showSequenceEnd:
        plt.plot(0,0,color=colors[3], label='sequence end mark',linewidth=10)        
        
    plt.legend(loc=4)
            
    wavPlayer(samples,samplerate)
```

```python
def similarity(i,events):
    if (i==len(events)):
        return 1
    (sorted_distances_indices,distances) = similarityMatrix(i)
    return 1-distances[i+1]
```

**Beispiel 3a mit similarity-Werten**

```python
plotSimilarityWithPlayer(sonicevents[0:10],audio,showSimilarity=1)
```

Wie bei *proximity* soll die Gruppengrenze anhand der Ähnlichkeit durch einen Schwellenwert festgelegt werden. Betrachtet wird dazu der *similarity*-Vektor zum nachfolgenden Klangereignis normalisiert im Verhältnis zu allen anderen Klangereignissen.

$$sequenceEnd(i)=\left\{\begin{matrix} 
 1 & \text{falls }  similarity(i) < t\\
 0 & \text{falls }  similarity(i) \geq  t\\
\end{matrix}\right.$$

```python
def sequenceEnd(i,events,t=0.5):
    if similarity(i,sonicevents)<t:
        return 1
    if similarity(i,sonicevents)>= t:
        return 0
    return 1
```
