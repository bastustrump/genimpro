```python
def flattenPhenotype(phenotype):
    flattend = {}
    for attribute in phenotype:
        for value in phenotype[attribute]:
            flattend[attribute + " " + value] = phenotype[attribute][value]
    return flattend
```

```python
def plotPhenotypeAttributes(phenotypes):
    values=[]
    for phenotype in phenotypes:
        values.append(flattenPhenotype(phenotype))

    attributes = {}
    colors = cm.rainbow(np.linspace(0, 1, len(values[0])))
    colorindex = 0
    
    for attribute in values[0]:
        attributes[attribute] = [values[i][attribute] for i in range(0,len(values))]
        
        fig = plt.figure(figsize=(20, 10), dpi=300)
        plt.plot(attributes[attribute],label=attribute,color=colors[colorindex])
        plt.legend()
        plt.show()
        colorindex+=1
```

```python
plotPhenotypeAttributes(phenotypes)
```
